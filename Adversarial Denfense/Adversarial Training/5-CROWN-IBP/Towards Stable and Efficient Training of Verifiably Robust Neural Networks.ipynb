# Verification Specifications<br />验证的规范



Each element in vector $m := Cf (x) \in \mathbb{R}^{n_{L}}$ gives us margins between class $y$ and all other classes.  
向量$m := Cf (x) \in \mathbb{R}^{n_{L}}$中的每个元素都给出了类$y$和所有其他类之间的边界。

We define the lower bound of $m$ for all $x \in S(x_{k}, \epsilon)$ as $\underline{m}(x_{k}, \epsilon)$.When all elements of $\underline{m}(x_{k}, \epsilon)$ > 0, $x_{k}$ is verifiably robust for any perturbation with $l_{∞}$ norm less than $\epsilon$.  
我们将所有$x \in S(x_{k}, \epsilon)$的$m$的下界定义为$\underline{m}(x_{k}, \epsilon)$。当$\underline{m}(x_{k}, \epsilon)$的所有元素都大于0时，$x_{k}$对于任何$l_{∞}$范数小于$\epsilon$的扰动都是可验证的鲁棒的。
