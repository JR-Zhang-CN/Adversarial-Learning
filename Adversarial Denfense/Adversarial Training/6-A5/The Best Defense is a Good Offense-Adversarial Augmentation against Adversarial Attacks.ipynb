# [The Best Defense is a Good Offense: Adversarial Augmentation against Adversarial Attacks<br />最好的防御是好的进攻：对抗增强防御](https://ieeexplore.ieee.org/document/10204336/)

- We introduce A5 (Adversarial Augmentation Against Adversarial Attacks), a comprehensive framework for preemptive augmentation to make data and physical objects certifiably robust against MitM adversarial attacks. As far as we know, this is the first time certified defense is achieved in this way.  
我们介绍了 A5（对抗攻击性增强），这是一个全面的先发制人增强框架，可使数据和物理对象具有经认证的鲁棒性，可抵御 MitM 对抗性攻击。据我们所知，这是首次以这种方式实现认证防御。 
- We study the connection between the robustness of the legacy classifier, the magnitude of the defensive augmentation, and the protection level delivered by A5. We show A5 achieving state-of-the-art certified protection against MitM attacks by training a robustifier DNN coupled with the legacy classifier, and even better results for co-training of the robustifier and classifier. We perform a critical, visual inspection of the robustified images to answer an interesting theoretical question (how does a non-attackable image look like?) and potentially provide directions for the acquisition of inherently robust images.  
我们研究了传统分类器的鲁棒性、防御增强的幅度和 A5 提供的保护级别之间的联系。我们展示了 A5 通过训练与传统分类器相结合的鲁棒性 DNN，实现了最先进的针对 MitM 攻击的认证保护，并且在鲁棒性分类器和分类器的联合训练中取得了更好的效果。我们对经过健壮处理的图像进行了严格的目视检查，以回答一个有趣的理论问题(不可攻击的图像是什么样的?)，并可能为获取固有鲁棒图像的方向提供指导。 

# Three main philosophies for the development of defenses against adversarial attacks<br />对抗防御的三个主要理念

1. Adversarial training. Attacks are generated while training the DNN. Certified defense can be seen as a SOTA refined form of adversarial training.  
对抗训练。在训练 DNN 时生成攻击。认证防御可以看作是对抗训练的 SOTA 精细形式。
2. Active countermeasures taken after the attack is crafted. Image purification projects an attacked image onto the manifold of natural images through a generative model. Randomization can also be used after the attack.  
被攻击后采取的主动对策。图像净化通过生成模型将被攻击的图像投影到自然图像的流形上。随机化也可以在攻击之后使用。
3. A5 belongs to a third class of methods, that put in place countermeasures before the attack.  
A5 属于第三类方法，即在攻击之前采取对策。



The closest algorithm to $A^{5}$ is preemptive robustification: given an image $x$ and a classifier $C$ that can be fine-tuned, it computes $y_{0} = C(x)$ and randomly samples attacks $δx_{A}$ against $x+δx_{D}$ to find a defensive perturbation $δx_{D}$, such that $(x + δx_{D} + δx_{A}) ∈ y_{0}$. $A^{5}$ uses certified bounds.  
最接近 $A^{5}$ 的算法是先发制人的强化：给定一个图像 $x$ 和一个可以微调的分类器 $C$，它计算 $y_{0} = C(x)$ 并随机采样攻击 $δx_{A}$ 对 $x+δx_{D}$，以找到防御扰动 $δx_{D}$，使得 $(x + δx_{D} + δx_{A}) ∈ y_{0}$。$A^{5}$ 使用认证边界。

The second idea behind $A^{5}$ comes from physical adversarial examples. Reverting the attacker and defender roles, the complementary cost function leads to the creation of unadversarial examples: patches or textures that, once applied to a real 3d object, reinforce (instead of degrading) the DNN desired behaviour.  
$A^{5}$ 背后的第二个想法来自物理对抗性示例。将攻击者和防御者的角色恢复到原来的状态，互补的成本函数导致了非对抗样本的创建：补丁或纹理，一旦应用于真实的 3d 对象，就会加强(而不是降低) DNN 的期望行为。
