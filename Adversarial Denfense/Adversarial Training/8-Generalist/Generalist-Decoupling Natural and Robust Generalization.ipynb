# [Generalist: Decoupling Natural and Robust Generalization<br />Generalist: 解耦自然泛化能力和鲁棒泛化能力](https://ieeexplore.ieee.org/document/10205177/)

Previous works all focus on the data used for training while we propose to tackle the tradeoff problem from the perspective of the training paradigm in this paper.  
以前的工作都集中在用于训练的数据上，而我们在本文中建议从训练范式的角度来解决权衡问题。

Specifically, for each sub-task, we train a base learner on natural/adversarial datasets with the task-specific configuration while sharing the same model architecture. The parameters of base learners are collected and combined to form a global learner at intervals during the training process, which is then distributed to base learners as initialized parameters for continued training. We name the framework as Generalist.  
具体来说，对于每个子任务，我们使用特定于任务的配置在自然/对抗数据集上训练基学习器，同时共享相同的模型架构。在训练过程中，收集并组合基学习器的参数，形成全局学习器，然后将其分发给基学习器作为继续训练的初始化参数。我们将该框架命名为Generalist。

### Meta-Learning<br />元学习

Suppose $A$ is divided into non-overlapping splits $V$ and $W$, the model is first trained on the training sets and then guided by a small validation set on a set of tasks to make the trained model can be well adapted to new tasks.  
假设$A$被分为不重叠的拆分$V$和$W$，模型首先在训练集上进行训练，然后在一组任务的小验证集上进行指导，使得训练的模型可以很好地适应新任务。
