# [Towards Transferable Targeted Adversarial Examples<br />可迁移有目标的对抗样本](https://ieeexplore.ieee.org/document/10204634/)

- We propose a general framework for targeted adversarial attack that works in **both non-cross-domain and cross-domain scenarios**. This framework could be easily integrated with other targeted adversarial attacks to improve their cross-model and cross-domain targeted transferability.  
我们提出了一个适用于**非跨域和跨域场景**的定向对抗攻击通用框架。这个框架可以很容易地与其他有针对性的对抗攻击相结合，以提高其跨模型和跨领域的针对性可转移性。

- Our proposed Transferable Targeted Adversarial Attack could extract the target class distribution from **feature-wise and label-wise** levels to promote perturbations to acquire label-related and transferable distribution information of the target class, thus generating highly transferable targeted adversarial examples.  
我们提出的可转移的有目标的对抗攻击可以从**特征和标签**两个层面提取目标类别的分布，以促进扰动获取与标签相关的和可转移的目标类别的分布信息，从而生成高度可转移的有目标的对抗样本。

Existing works have limited transferability because they only utilize feature or label information and cannot adequately characterize the target class information.  
现有的工作的可转移性有限，因为它们只利用特征或标签信息，不能充分表征目标类信息。



Existing transferable adversarial attacks focusing on the cross-model transferability of adversarial examples make a strong assumption that the domain of the target model is accessible for the attackers to perform adversarial attacks in non-cross-domain scenarios, which is not applicable in more practical cross-domain scenarios.  
现有的可转移对抗攻击侧重于对抗样本的跨模型可转移性，做出了一个强有力的假设，即目标模型的域是攻击者在非跨域场景中进行对抗性攻击时可以访问的，这不适用于更实际的跨域场景。

## Our Method<br />我们的方法



### Label-Feature Dual Discrimination<br />标签特征双重判别

With the input of the image from adversarial example $x^{adv}_{s}$ and the target class example $x_{t}$, the label discriminator outputs the predicted corresponding label $D_{ψ}(x^{adv}_{s})$ and $D_{ψ}(x_{t})$. Moreover, to further leverage the feature information from $x_{t}$ and $x_{t}$, we obtain the corresponding feature map $r^{adv}_{s}$ and $r_{t}$ from intermediate layer outputs of $D_{ψ}$.  
对于来自对抗样本$x^{adv}_{s}$和目标类别样本$x_{t}$的图像输入，标签判别器输出预测的相应标签$D_{ψ}(x^{adv}_{s})$和$D_{ψ}(x_{t})$。此外，为了进一步利用$x_{t}$和$x_{t}$的特征信息，我们从$D_{ψ}$的中间层输出中获得相应的特征图$r^{adv}_{s}$和$r_{t}$。





The feature discriminator is a one-class classifier, where the feature maps $r^{adv}_{s}$ and $r_{t}$ extracted by the label discriminator serve as inputs and the probabilities of the feature maps $r^{adv}_{s}$ and $r_{t}$ being from the target class are outputs, i.e., $D_{ξ}(r^{adv}_{s})$ and $D_{ξ}(r_{t})$. The output of 1 from $D_{ξ}$ indicates that the feature map is from the target class image, and an output of 0 means the opposite.  
特征判别器是一个单类分类器，其中由标签判别器提取的特征图$r^{adv}_{s}$和$r_{t}$作为输入，特征图$r^{adv}_{s}$和$r_{t}$来自目标类的概率作为输出，即$D_{ξ}(r^{adv}_{s})$和$D_{ξ}(r_{t})$。$D_{ξ}$的输出为1表示特征图来自目标类图像，输出为0表示相反。



### Random Perturbation Dropping (RPD)<br />随机扰动丢弃



It is worth noting that there may exist a case where the adversarial perturbations generated by the trained generator **are concentrated in a certain region of the image**, which **decreases the robustness of the distribution information of the target class** on adversarial examples and compromises the transferability. Instead, to ensure high transferability, we need to render the perturbations that matter uniformly distributed over the entire image.  
值得注意的是，可能存在这样一种情况，即训练生成器生成的对抗扰动**集中在图像的某个区域**，这**降低了对抗样本上目标类的分布信息的鲁棒性**，并损害了可转移性。相反，为了确保高可转移性，我们需要使重要的扰动均匀分布在整个图像上。

### Generative Adversarial Training<br />生成对抗训练

We generalize the adversarial example generation as an optimization problem, for which we design the loss function $L_{G}$ and $L_{D}$ for the generator and the feature discriminator, respectively. By minimizing the loss functions, we iteratively update the parameters $θ$ of the generator $G_{θ}$ and the parameters $ξ$ of the feature discriminator $D_{ξ}$ via gradient descent.  
我们将对抗样本生成泛化为一个优化问题，为此我们分别为生成器和特征判别器设计了损失函数$L_{G}$和$L_{D}$。通过最小化损失函数，我们通过梯度下降迭代地更新生成器$G_{θ}$的参数$θ$和特征判别器$D_{ξ}$的参数$ξ$。
